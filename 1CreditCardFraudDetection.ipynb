```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split 

```

```python
y = df["Class"].values
x = df.drop("Class", axis=1).values
```

```python
smote = SMOTE(random_state=42)
x_resampled , y_resampled = smote.fit_resample(x,y) ## this will make minority index and data to be in model more often .
```

```python
x_train , x_test, y_train , y_test = train_test_split(x_resampled , y_resampled , random_state=25 , test_size=0.12)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation , Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlyStopping = EarlyStopping(mode="min" , monitor = "val_loss" , verbose= 1 , patience = 25 )
```

```python
model = Sequential()

model.add(Dense(units=29 , activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=12 , activation="relu"))
model.add(Dropout(0.5))


model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam" , loss= "binary_crossentropy" , metrics = ["accuracy"] )
```

```python
model.fit(x=x_train  , y = y_train , epochs=150 , verbose= 1 , batch_size=1024 , validation_data=(x_test, y_test) , callbacks=[earlyStopping])
```

```python
model.history.history
```

```python
lossData = pd.DataFrame(model.history.history)
```

```python
lossData.plot()
plt.show()
```

```python
yPred = model.predict(x_test)
```

```python
yPred 
```

```python
classed_result = (yPred > 0.3).astype(int)
```
